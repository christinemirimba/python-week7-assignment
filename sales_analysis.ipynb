```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inline
%matplotlib inline

# Set a nice style for plots
sns.set_style("whitegrid")

# Load dataset
try:
    df = pd.read_csv("supermarket_sales.csv")
    print("Dataset loaded successfully!\n")
except FileNotFoundError:
    print("supermarket_sales.csv not found. Please check the file path.")

# Explore dataset
# Display first few rows
df.head()

# Dataset info
df.info()

# Missing values
df.isnull().sum()

# Data Cleaning
# Drop rows with missing values (if any)
df = df.dropna()
print("Data cleaned! No missing values remain.")

# Basic Analysis
# Statistical summary
df.describe()

# Average Total Sales per Branch
df.groupby("Branch")["Total"].mean()

# Average Rating per Product line
df.groupby("Product line")["Rating"].mean()

# Visualization
# Line chart (Total Sales Over Time)
df["Date"] = pd.to_datetime(df["Date"])
sales_over_time = df.groupby("Date")["Total"].sum()

plt.figure(figsize=(10, 5))
sales_over_time.plot()
plt.title("Total Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()

# Bar chart (Average Sales by Product Line)
plt.figure(figsize=(8, 5))
df.groupby("Product line")["Total"].mean().plot(kind="bar", color="skyblue")
plt.title("Average Total Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Average Sales")
plt.show()

# Histogram (Distribution of Ratings)
plt.figure(figsize=(8, 5))
df["Rating"].plot(kind="hist", bins=20, color="lightgreen", edgecolor="black")
plt.title("Distribution of Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()

# Scatter Plot (Unit Price vs Quantity)
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Unit price", y="Quantity", data=df, hue="Branch")
plt.title("Unit Price vs Quantity")
plt.xlabel("Unit Price")
plt.ylabel("Quantity")
plt.legend(title="Branch")
plt.show()








```
